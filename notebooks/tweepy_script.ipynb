# Run the tweepy script to save tweets that include the desired search query to a json file on disk.

Download 4-5% of each file IDs.

```python
filenames=['election-filter1.txt', 'election-filter2.txt', 'election-filter3.txt', 'election-filter4.txt',
           'election-filter5.txt', 'election-filter6.txt'] 
    # 'first_debate.txt', 'election-day.txt', 'democratic-convention-filter.txt', 'republican-convention-filter.txt', 
    # 'second-debate.txt', 'third-debate.txt', 'vp-debate.txt'
```

```python
import sys
import jsonpickle
import os
import numpy as np
import time

ids_dir='non_troll_ids'
save_dir='non_troll_twts'
pct_from_file=0.05
tweetsPerQry=100  # this is the max the API permits
```

```python
for f in filenames:
    print('On file: {}'.format(f))
    
    with open(os.path.join(ids_dir, f), 'r') as id_file:
        all_ids = id_file.read().splitlines()
    
    # select 5% randomly
    # rand_ids = list(np.random.choice(all_ids, size=int(pct_from_file*len(all_ids))))
    rand_ids = list(np.random.choice(all_ids, size=100000, replace=False))
    
    num_requests=int(len(rand_ids)/tweetsPerQry) + 1
    out_file = f.replace('.txt', '')+'-tweets.json'
    tweetCount = 0

    # iterate over requests and write to file
    # 900 requests per 15 mins => 60 requests per min => sleep 1 second
    with open(os.path.join(save_dir, out_file), 'w') as out:    
        for i in range(num_requests):
            try:
                new_tweets = api.statuses_lookup(rand_ids[i*tweetsPerQry:(i+1)*tweetsPerQry])
                for tweet in new_tweets:
                    out.write(jsonpickle.encode(tweet._json, unpicklable=False) + '\n')
                tweetCount += len(new_tweets)
                print("Downloaded {0} tweets".format(tweetCount))
            except tweepy.TweepError:
                print('tweepError on iter {} of {}'.format(i, num_requests))
                continue
            time.sleep(1)
```
